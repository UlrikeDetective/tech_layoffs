```python
import pandas as pd

df = pd.read_csv('/Users/ulrike_imac_air/projects/DataScienceProjects/tech_layoffs_project/tech_layoffs_csv/tech_layoffs_Q2_2024.csv')

# Display the first few rows of the DataFrame
display(df.head())

# Display the DataFrame information
display(df.info(verbose=True))
```

```python
import pandas as pd

# Group by "Location_HQ" and sum the "Laid_off" entries
df_grouped = df.groupby('Location_HQ')['Laid_Off'].sum().reset_index()

# Rename the columns for clarity
df_grouped.columns = ['Location_HQ', 'Total_Laid_off']

print(df_grouped)

```

```python
df_grouped.to_csv('layoffs_HQ_sum.csv', index=False)
```

```python
import pandas as pd

df_country = df.groupby('Country')['Laid_Off'].sum().reset_index()

# Rename the columns for clarity
df_country.columns = ['Country', 'Total_Laid_off']

print(df_country)
```

```python
df_country.to_csv('layoffs_country_sum.csv', index=False)
```

```python
import pandas as pd

df_continent = df.groupby('Continent')['Laid_Off'].sum().reset_index()

# Rename the columns for clarity
df_continent.columns = ['Continent', 'Total_Laid_off']

print(df_continent)
```

```python
df_continent.to_csv('layoffs_continent_sum.csv', index=False)
```

```python
# Group by "Location_HQ" and "Year" and sum the "Laid_Off" entries
# Group by "Location_HQ" and "Year" and sum the "Laid_Off" entries
df_HQ_Y = df.groupby(['Location_HQ', 'Year'])['Laid_Off'].sum().reset_index()

# Pivot the DataFrame to get years as columns
df_HQ_Y_pivot = df_HQ_Y.pivot(index='Location_HQ', columns='Year', values='Laid_Off').fillna(0)

# Add a 'Total_Laid_off' column
df_HQ_Y_pivot['Total_Laid_off'] = df_HQ_Y_pivot.sum(axis=1)

# Ensure all relevant years are present in the columns
for year in range(2020, 2025):
    if year not in df_HQ_Y_pivot.columns:
        df_HQ_Y_pivot[year] = 0

# Reorder columns to have years first and then 'Total_Laid_off'
columns_order = list(range(2020, 2025)) + ['Total_Laid_off']
df_HQ_Y_final = df_HQ_Y_pivot[columns_order].reset_index()

print(df_HQ_Y_final)
```

```python
df_HQ_Y_final.to_csv('layoffs_HQ_sum_year.csv', index=False)
```
